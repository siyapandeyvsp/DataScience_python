```python
turtle=cv2.imread('turtles.jpg') #bgr by default 
```

```python
plt.imshow(turtle)
plt.axis('off') #to remove pixel notation on x nd y axis
plt.show()
```

```python
turtle_color_converted=cv2.cvtColor(turtle,cv2.COLOR_BGR2RGB)
plt.imshow(turtle_color_converted)
plt.axis('off')
```

```python
turtle_color_gray=cv2.cvtColor(turtle,cv2.COLOR_BGR2GRAY)
plt.imshow(turtle_color_gray,cmap='gray')
plt.axis('off')
```

```python
plt.imshow(np.hstack([turtle,turtle_color_converted]))
plt.axis('off')
```
